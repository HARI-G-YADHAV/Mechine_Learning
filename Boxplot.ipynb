```python
import pandas as pd
import random
import datetime
```

```python
num=30
dates=pd.date_range(start='01-01-2003',end='01-01-2004',periods=num)
price=[round(random.uniform(1,100),2) for i in range(num)]
```

```python
product_id=[random.randint(1,100) for i in range(num)]
```

```python
quantity_purchased=[random.randint(1,100) for i in range(num)]
```

```python
serial_no=[f'SN-{random.randint(1,100)}' for i in range(num)]
```

```python
user_id=['U' + str(random.randint(1,100)) for i in range(num)]
```

```python
user_type=['L','U']
user_class=['A','B','C']
purchase_week=[date.isocalendar()[1] for date in dates]
```

```python
data={
    'Date':dates,
    'Price':price,
    'Product':product_id,
    'Quantity':quantity_purchased,
    'Serial':serial_no,
    'User_id':random.choices(user_id,k=num),
    'User_type':random.choices(user_class,k=num),
    'Purchase':purchase_week
}
```

```python
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['Quantity'])
plt.xlabel('Quantity')
plt.title('quantity')
```

```python
plt.boxplot(df['Purchase'])
plt.xlabel('Quantity')
plt.title('quantity')
```
